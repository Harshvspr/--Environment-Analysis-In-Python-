# 🌍 Environmental Analysis of Manesar (Gurgaon)

##  Introduction

Manesar is a fast-growing industrial town in the Gurugram district of Haryana. With the rapid expansion of IMT phases, the region faces multiple environmental challenges such as:

- Air pollution  
- Water stress  
- Noise pollution  
- Land use imbalance  

This project analyzes key environmental factors in Manesar using real datasets and official reports. The objective is to assess the impact of industrial growth on:

- Air quality  
- Groundwater conditions  
- Rainfall patterns  
- Noise levels  
- Water demand and supply  
- Land use changes  

---
>   **Note:** Most of the environmental datasets used in this analysis  
> (rainfall, groundwater, noise, land use) are based on official reports from 2019 and 2024.  
>  
> Due to limited public availability of real-time data for Manesar,  
> recent proxy datasets from Gurgaon (weather, AQI) are used to supplement the analysis.

### so there is no null value here , now we have to move  directly on charts, what insights we get from the chart?


```python
plt.figure(figsize=(12,5))
plt.plot(df_aqi["Time"],df_aqi["PM2.5 (µg/m³)"],label="PM2.5",marker = 'o')# time vs pollution conclude moring and night have less pollution.
plt.plot(df_aqi["Time"], df_aqi["PM10 (µg/m³)"], label="PM10", marker='o', linestyle='--')
plt.title("24-Hour Particulate Matter Concentration – Manesar")
plt.xlabel("Time")
plt.ylabel("µg/m³")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_aqi["Time"],df_aqi["AQI"],color = "green",label = "AQI",marker = 'o')
plt.title("Time vs AQI graph")
plt.xlabel("AQI vlaue")
plt.ylabel("Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation = 45);
# here we see how AQI vary with time, NOTE --> at the mid time around 1pm AQI reach its highest .
```

### now we explore manesar history data with concentration,and how pollutions pm lvl vary in manesar area.

```python
# so here to access all columns and create graph we have to use bar graph.
columns = ["PM2.5 (ug/m3)","PM10 (ug/m3)","NO (ug/m3)","NO2 (ug/m3)","NOx (ppb)","NOx (ppb)","SO2 (ug/m3)"]
avg_value = df_AQI[columns].mean()
avg_value.plot(kind = 'bar',figsize=(10,5),color = 'skyblue')
plt.title("Manesar, PM-Level")
plt.xlabel("Concentration")
plt.ylabel("Pollutant")
plt.xticks(rotation=(45));
```

```python
# now we create heatmap to see is there any correlation among these pollutant gas or not .
PM = [ "PM2.5 (ug/m3)", "PM10 (ug/m3)", "NO (ug/m3)", "NO2 (ug/m3)",
    "NOx (ppb)", "NH3 (ug/m3)", "SO2 (ug/m3)", "CO (mg/m3)",
    "Ozone (ug/m3)", "Benzene (ug/m3)", "Toluene (ug/m3)",
       ]
corr_matrix = df_AQI[PM].corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot= True,cmap = 'coolwarm')
plt.title("Pollution which is harmful when it combines")
```

##  Final Conclusion

This project aimed to understand the environmental situation in the Manesar region by analyzing air quality and land use patterns. Through a mix of data visualizations and simple analysis, we looked at both pollution levels and how land has been allocated over different phases of development.

- The data showed that **PM2.5 and PM10 levels** were quite high on most days, pointing toward serious air pollution in the area.
- In the **Rainfall vs PM2.5 graph**, we noticed that pollution levels tend to drop when there’s rainfall. This shows that rain plays a role in cleaning the air, at least temporarily.
- The **correlation heatmap** helped us identify which pollutants are closely related. For example, PM2.5, PM10, NO2, and NOx had strong correlations, likely due to traffic and industrial sources.
- The **stacked bar chart** of land use phases showed that most of the land was used for industrial and residential purposes. Areas like parks and public utilities got much less space, especially in the newer phases.

Overall, this analysis highlights how fast industrial growth is affecting both air quality and city planning in Manesar. The insights from this project can be helpful for future planning decisions that aim to balance development with environmental health.
