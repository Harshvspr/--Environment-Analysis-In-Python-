##  Introduction

Manesar is a fast-growing industrial town in the Gurugram district of Haryana. With the rapid expansion of IMT phases, the region faces multiple environmental challenges such as:

- Air pollution  
- Water stress  
- Noise pollution  
- Land use imbalance  

This project analyzes key environmental factors in Manesar using real datasets and official reports. The objective is to assess the impact of industrial growth on:

- Air quality  
- Groundwater conditions  
- Rainfall patterns  
- Noise levels  
- Water demand and supply  
- Land use changes  

---
>   **Note:** Most of the environmental datasets used in this analysis  
> (rainfall, groundwater, noise, land use) are based on official reports from 2019 and 2024.  
>  
> Due to limited public availability of real-time data for Manesar,  
> recent proxy datasets from Gurgaon (weather, AQI) are used to supplement the analysis.
